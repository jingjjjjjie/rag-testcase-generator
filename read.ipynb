## 示例 3: 用于 Prompt 模板

````python
# 准备上下文
context_dict = {
    "passage_1": "Artificial Intelligence is transforming the tech industry.",
    "passage_2": "Machine learning models require large amounts of training data."
}

# 格式化为编号文档
formatted_context = list_to_docided_string(context_dict)

# 创建 prompt
prompt_template = """Given the following context:

[[CONTEXT]]

Question: [[QUESTION]]
Answer: [[ANSWER]]
"""

prompt = prompt_template.replace('[[CONTEXT]]', formatted_context)
prompt = prompt.replace('[[QUESTION]]', "What is AI used for?")
prompt = prompt.replace('[[ANSWER]]', "AI is used for transforming industries.")

print(prompt)
````

## 输出格式说明

`list_to_docided_string` 函数会将字典转换为以下格式：

```
0. <doc>
    <doc-name>doc_id</doc-name>
    <detailed-desc>doc_content</detailed-desc>
</doc>
1. <doc>
    <doc-name>doc_id</doc-name>
    <detailed-desc>doc_content</detailed-desc>
</doc>
```

这种格式便于 LLM 理解和引用文档。
